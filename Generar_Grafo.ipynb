Ahora comienza la creación de nuestros datos almacenes
Luego lo convertiremos ha archivo csv y de igual manera haremos
Con los puntos entrega y se hara de acuerdo al requisito que se nos piden

Almacenes entre 50 y 100 puntos de distribución (almacenes).

Puntos de entrega entre 2500 y 5000 puntos de entrega.

```python
def Almacenes_():
  A = random.randint(50,100)
  Almacenes = []

  cx_A = np.random.randint(1,1000, size=A)
  cy_A = np.random.randint(1,1000, size=A)

  for u in range(A):
    Almacenes.append((cx_A[u], cy_A[u]))

  return Almacenes
```

```python
Almacenes_()
```

```python
def Entregas_():
  E = random.randint(2500,5000)
  Entregas = []
  cx_E = np.random.randint(1,1000, size=E)
  cy_E = np.random.randint(1,1000, size=E)
  for v in range(E):
    Entregas.append((cx_E[v],cy_E[v]))

  return Entregas
```

```python
Entregas_()
```

Ahora convertiremos los datos generados ha archivos CSV tanto
de los almacenes y puntos de entrega

Primeroharemos un ejemplo de como se escribe un archivo
mediante la función csv.writer y para eso importaremos la librería
csv

```python
import csv

#Primero se importara el módulo csv, y la función writer() creará un objeto apto
#para la escritura.

#Para iterar los datos sobre las filas, necesitaremos usar la función writerows()

Datos = [(417, 638),
         (850, 611),
         (630, 615),
         (655, 712),
         (369, 352),
         (897, 922),]

Generar = open('Ejemplo.csv','w')
with Generar:
  writer = csv.writer(Generar)
  writer.writerows(Datos)

print("Escritura completada")
```

En esta imagen se puede observar que ya se creo el archivo csv con nuestros datos de igual manera haremos para los almacenes y puntos de entrega.

```python
Resultado1 = Almacenes_()
Resultado2 = Entregas_()
```

```python
alma = open('almacenes.csv','w')
with alma:
  writer = csv.writer(alma)
  writer.writerows(Resultado1)

entre = open('puntos_entrega.csv','w')
with entre:
  writer = csv.writer(entre)
  writer.writerows(Resultado2)
```

Muy bien estos archivos generados tanto de Almacenes y Puntos de entrega lo utilizaremos para crear nuestro grafo

Con la función draw lo dibujaremos mediante las coordenas que se genero en el arreglo de par de tuplas donde las x se representan
como la primera columna e y representan la segunda columnas

```python
def DRAW():
  #Se crea el objeto figura y se establece el ancho de la figura de 20 pulgadas y la altura
  #de 20 pulgadas
  plt.figure(num=1,figsize=(15,15),facecolor='white')
  plt.scatter(cx_A,cy_A,s=30,color="green",label="Almacenes")
  plt.scatter(cx_E,cy_E,s=30,color="yellow",label="Entregas")

  plt.xlabel("X")
  plt.ylabel("Y")
  plt.title("Representación de los almacenes y puntos de entrega en una ciudad rectangular de 1000x1000")
  plt.legend()
  
  return plt.show()
```

```python
def spell_csv(headers, fields, filename):
  with open(filename, 'w') as f:   
    write = csv.writer(f)
    if headers != None: write.writerow(headers)
    write.writerows(fields)
def interpret_csv(filename):
  csv_file = list()
  with open(filename) as File:
    reader = csv.reader(File, delimiter=',', quotechar=',', quoting=csv.QUOTE_MINIMAL)
    for row in reader:
        if row[0] != 'x':
          row[0], row[1] = int(row[0]), int(row[1])
          csv_file.append(row)
  csv_file.sort(key = lambda x: (x[0], x[1]))
  return csv_file
warehouse = interpret_csv("almacenes.csv")
Home = interpret_csv("puntos_entrega.csv")
x, y = np.array(warehouse).T
x2, y2 = np.array(Home).T
plt.figure(figsize=(15, 15))
plt.scatter(x,y, 2, c="red")
plt.scatter(x2,y2, 2, c="blue")
plt.show()
for i, _ in enumerate(warehouse):
  warehouse[i].append("W")
for i, _ in enumerate(Home):
  Home[i].append("H")
city = list()
city.extend(warehouse)
city.extend(Home)
city.sort(key = lambda x: (x[0], x[1]))
graph = [[] for _ in city]
for i, _ in enumerate(city):
  for j, _ in enumerate(city):
    if i == j: continue
    if city[i][0] == city[j][0] or city[i][1] == city[j][1]: graph[i].append(j)
spell_csv(['x', 'y', 'type'], city, 'town.csv')
spell_csv(None, graph, 'graph.csv')
```
